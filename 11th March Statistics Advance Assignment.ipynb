```python
#Q4: Explain Bayes's theorem with an example.
'''
Bayes' theorem is a mathematical formula that is used to update our beliefs about the probability of an event, given new evidence.

The formula for Bayes' theorem is:
P(A|B) = P(B|A) * P(A) / P(B)

For example, let's say that we know that 10% of people have a certain disease. We also know that a certain test is 90% accurate 
at detecting the disease. If we take the test and it comes back positive, what is the probability that we actually have the disease?

We can use Bayes' theorem to answer this question.
In this case, event A is "having the disease" and event B is "testing positive for the disease." 
We know that P(A) = 0.1, P(B|A) = 0.9, and P(B) = 0.1. We can plug these values into the formula to get:
P(A|B) = P(B|A) * P(A) / P(B)
= 0.9 * 0.1 / 0.1
= 0.9

This means that the probability of having the disease, given that we tested positive, is 90%
'''
```

```python
#Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
'''
A confidence interval is a range of values that is likely to include an unknown population parameter. 
Confidence intervals are used to estimate the value of a population parameter, such as the mean or standard deviation, from a sample.

The confidence interval is calculated by adding and subtracting a margin of error to the sample statistic. 
The margin of error is calculated using the standard error of the statistic and the confidence level.

Confidence interval = sample mean ± margin of error

Margin of error is calculated using the following formula:
Margin of error = z * standard error of the mean

Standard error of the mean = σ / √n

For example, let's say that we have a sample of 100 people and the mean height is 6 feet. 
The standard deviation of the population is 2 inches. We want to calculate a 95% confidence interval for the mean height.

The z-score for a 95% confidence interval is 1.96. The standard error of the mean is 0.2 inches.

The confidence interval is calculated as follows:

Confidence interval = 6 feet ± 1.96 * 0.2 inches
= 6 feet ± 0.39 inches
= 5.61 inches to 6.39 inches

This means that we are 95% confident that the true mean height of the population is between 5.61 inches and 6.39 inches.
'''
```

```python
#Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
#Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
'''

The margin of error in a confidence interval is the amount of uncertainty that is associated with the interval. 

It is calculated as follows:

Margin of error = z * standard error of the mean

where:
z is the z-score for the desired confidence level. For a 95% confidence interval, z = 1.96.
Standard error of the mean is calculated using the following formula:
Standard error of the mean = σ / √n

The margin of error decreases as the sample size increases. This is because a larger sample size is more likely to be 
representative of the population, and therefore less likely to be affected by random variation.

For example, let's say that we want to estimate the mean height of all adults in the United States. 
We could take a sample of 100 adults and calculate the mean height of the sample. 
The margin of error for this estimate would be relatively large, 
because a sample of 100 adults is not very representative of the entire population of the United States.

However, if we took a sample of 10,000 adults, the margin of error would be much smaller. 
This is because a sample of 10,000 adults is more likely to be representative of the population, 
and therefore less likely to be affected by random variation.
'''
```

```python
#Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.
'''
The z-score is calculated as follows:

z = (x - μ) / σ

where:

x is the data point value
μ is the population mean
σ is the population standard deviation

In this case, x = 75, μ = 70, and σ = 5. Plugging these values into the formula, we get:
z = (75 - 70) / 5
= 1.00

The z-score of 1.00 tells us that the data point is 1 standard deviation above the mean. 
This means that the data point is higher than 68% of the data points in the population.

In general, a z-score of 1.00 is considered to be a relatively small effect size. 
This means that the data point is not very different from the mean.
'''
```

```python
# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
# of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
# significantly effective at a 95% confidence level using a t-test.
'''
Sample size = 50, Sample mean = 6, Sample standard devaition = 2.5, Confidence level = 0.95, Significance level = 0.05

Null hypothesis: Drug is significantly effective.
Alternate hypothesis: Drug is not significantly effective.

Degree of freedom (DOF) = 50 - 1 = 49

So, critical value for significant level of 0.05 = 2.0096

If t-value is less than -2.0096 or t-value is greater than 2.0096 then we will reject the null hypothesis else accept it.

t = (6 - 0) / (2.5 / sqrt(50))
  = 6 / (2.5 / 7.071)
  = (6 * 7.071) / 2.5
  = 42.426 / 2.5
  = 16.97
  
Since t-value > 2.0096. So we will reject the null hypothesis.

Hence there is no significant effect of drug.
'''
```

```python
#Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
#confidence interval for the true proportion of people who are satisfied with their job.
'''
Here are the steps on how to calculate the 95% confidence interval for the true proportion of people who are satisfied with their job:

1. Calculate the sample proportion.
2. Calculate the margin of error.
3. Add and subtract the margin of error to the sample proportion to get the confidence interval.

Calculate the sample proportion:

The sample proportion is calculated as follows:
p = x / n
where:
x is the number of people who are satisfied with their job
n is the sample size

In this case, x = 325 and n = 500. Plugging these values into the formula, we get:
p = 325 / 500
= 0.65

Calculate the margin of error:

The margin of error is calculated as follows:
ME = z * √(p * (1 - p) / n)
where:
z is the z-score for the desired confidence level. For a 95% confidence interval, z = 1.96.
p is the sample proportion
n is the sample size

In this case, z = 1.96, p = 0.65, and n = 500. Plugging these values into the formula, we get:

ME = 1.96 * √(0.65 * (1 - 0.65) / 500)
= 0.048

Add and subtract the margin of error to the sample proportion to get the confidence interval:

The confidence interval is calculated as follows:
CI = p ± ME

In this case, the confidence interval is:

CI = 0.65 ± 0.048
= 0.602 to 0.698


This means that we are 95% confident that the true proportion of people who are satisfied with their job is between 60.2% and 69.8%.
'''
```

```python
# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
# Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
# with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
# # significant difference in student performance using a t-test with a significance level of 0.01.
'''
Here are the steps on how to conduct a hypothesis test to determine if the two teaching methods have a significant difference 
in student performance using a t-test with a significance level of 0.01:

1. State the null and alternative hypotheses.
2. Calculate the test statistic.
3. Determine the p-value.
4. Make a decision.

**State the null and alternative hypotheses.**

The null hypothesis is that the two teaching methods have the same effect on student performance. 
The alternative hypothesis is that the two teaching methods have different effects on student performance.

**Calculate the test statistic.**

The t-test statistic is calculated as follows:

t = (x1 - x2) / (s1 / √n1 + s2 / √n2)

where:

x1 is the mean score of sample A
x2 is the mean score of sample B
s1 is the standard deviation of sample A
s2 is the standard deviation of sample B
n1 is the sample size of sample A
n2 is the sample size of sample B

In this case, x1 = 85, x2 = 82, s1 = 6, s2 = 5, n1 = 100, and n2 = 100. 
Plugging these values into the formula, we get:

t = (85 - 82) / (6 / √100 + 5 / √100)
= 0.71

Determine the p-value:

The p-value is the probability of obtaining a test statistic at least as extreme as the one that was observed, 
assuming that the null hypothesis is true. The p-value can be calculated using a t-table.

In this case, the p-value is 0.476. 
This means that there is a 47.6% chance of obtaining a t-statistic of 0.71 or more if the null hypothesis is true.

Make a decision:

Since the p-value is greater than the significance level of 0.01, we fail to reject the null hypothesis. 
This means that there is not enough evidence to conclude that the two teaching methods have different effects on student performance.
'''
```

```python
#Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
# of 65. Calculate the 90% confidence interval for the true population mean.
'''
Population mean = 60, Population Standard deviation = 8, Sample size = 50, Sample mean = 65, Confidence interval = 0.90
Hence, Significance level = 0.10

To calculate the 90% confidence interval for the true population mean, we can use the following formula:

CI = x̄ ± z*(s/√n)

where CI is the confidence interval, x̄ is the sample mean, z is the z-score corresponding to the desired confidence level (0.90), 
s is the sample standard deviation, and n is the sample size.

First, we need to find the z-score corresponding to the 90% confidence level. 
This can be done using a standard normal distribution table or calculator.
For a two-tailed test at α = 0.10 (since we want to find the middle 90% of the distribution), the z-score is 1.645.

Plugging in the given values, we have:

CI = 65 ± 1.645*(8/√50)

CI = 65 ± 2.313

The lower bound of the confidence interval is:

65 - 2.313 = 62.687

The upper bound of the confidence interval is:

65 + 2.313 = 67.313

Therefore, we can say with 90% confidence that the true population mean falls between 62.687 and 67.313.
'''
```

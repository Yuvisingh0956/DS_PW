```python
#Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.
import numpy as np
import scipy.stats as stats

null_hypothesis = "The population means are equal."
alternative_hypothesis = "The population means are not equal."

significance_level = 0.05

male_weights = np.array([180, 175, 185, 170, 175])
female_weights = np.array([160, 155, 165, 150, 155])

test_statistic = stats.ttest_ind(male_weights, female_weights)

critical_value = stats.t.ppf(significance_level, df=(len(male_weights) - 1) + (len(female_weights) - 1))

if np.any(test_statistic >= critical_value):
    print("The null hypothesis is rejected.")
else:
    print("The null hypothesis is not rejected.")
```

```python
#Q8. Define p-value and explain its significance in hypothesis testing.
'''
p-value is the probability of obtaining a test statistic at least as extreme as the one that was actually observed, under the assumption that the null hypothesis is true.

The p-value is used in hypothesis testing to determine whether to reject the null hypothesis. 
A p-value of less than or equal to the significance level (typically 0.05) indicates that the null hypothesis is likely to be false, and therefore the alternative hypothesis should be accepted.
A p-value greater than the significance level indicates that the null hypothesis is not likely to be false, and therefore the alternative hypothesis should not be accepted.
'''
```

```python
#Q10. Write a Python program to calculate the two-sample t-test for independent samples, 
#given two random samples of equal size and a null hypothesis that the population means are equal.
import numpy as np
import scipy.stats as stats

# Define the sample sizes.
n1 = 10
n2 = 10

# Generate the random samples.
sample1 = np.random.randint(0, 100, n1)
sample2 = np.random.randint(0, 100, n2)

# Calculate the means of the samples.
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)

# Calculate the standard deviations of the samples.
std1 = np.std(sample1)
std2 = np.std(sample2)

# Calculate the pooled standard deviation.
pooled_std = np.sqrt((std1**2 + std2**2) / (n1 + n2))

# Calculate the t-statistic.
t_statistic = (mean1 - mean2) / pooled_std

# Calculate the degrees of freedom.
df = n1 + n2 - 2

# Calculate the p-value.
p_value = stats.t.cdf(t_statistic, df)

# Print the results.
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
#Q12: What is t-statistic? State the formula for t-statistic.
'''

In statistics, a t-statistic is a statistical test statistic that is used to compare the means of two groups. 
It is calculated by dividing the difference between the means of the two groups by the standard error of the difference between the means. 

The formula for the t-statistic is as follows:
t = (μ1 - μ2) / SE(μ1 - μ2)
where:
μ1 is the mean of group 1
μ2 is the mean of group 2
SE(μ1 - μ2) is the standard error of the difference between the means
'''
```

```python
#Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
# sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
# Estimate the population mean revenue with a 95% confidence interval.
'''
The sample mean is $500 and the standard deviation is $50. 
The t-distribution critical value for a 95% confidence interval with 49 degrees of freedom is 2.009.

The confidence interval is calculated as follows:
(sample mean - t-critical value * standard deviation / sqrt(sample size), sample mean + t-critical value * standard deviation / sqrt(sample size))

On putting the data we get:
(500 - 2.009 * 50 / sqrt(50), 500 + 2.009 * 50 / sqrt(50))
= (459.45, 540.55)

Therefore, we can be 95% confident that the population mean revenue is between $459.45 and $540.55.
'''
```

```python
#Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
# standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
# is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
# pounds with a significance level of 0.01.

'''
Population mean = 5, Population Standard Deviation = 0.5, Sample mean = 4.8, Sample Size = 25, Significance level = 0.01

Null hypothesis: The true mean weight of the products is 5 pounds.
Alternative hypothesis: The true mean weight of the products is less than 5 pounds

z-test:
z = (sample mean - hypothesized mean) / (standard deviation/ sqrt(sample size)

So putting values in the equation:
z = (4.8 - 5) / (0.5 / sqrt(25))
  = (-0.2) / (0.5 / 5)
  = (-0.2 * 5) / 0.5
  = -1 / 0.5
  = -2
  
Area under the curve of -2 z-value is = 0.0228
Hence p-value = 0.0228

if p-value < significance value then we will reject null hypothesis else we will accept it.

Here, 0.0228 > 0.01. Hence we will accept the null hypothesis.

So, the true mean weight of the products is 5 pounds.
'''
```

```python
# Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
# 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
# score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
# groups are equal with a significance level of 0.01.
'''
For group 1: Sample mean = 80, Sample standard deviation = 10, Sample size = 30
For group 2: Sample mean = 75, Sample standard deviation = 8, Sample size = 40

Null hypothesis: The population means for the two groups are equal.
Alternate hypothesis: the population means for the two groups are not equal.

The test statistic is calculated as:

t = (x̄1 - x̄2) / sqrt[s1^2/n1 + s2^2/n2]

where x̄1 and x̄2 are the sample means, s1 and s2 are the sample standard deviations, and n1 and n2 are the sample sizes.

Substituting the given values, we have:

t = (80 - 75) / sqrt[(10^2/30) + (8^2/40)]
t = 2.85

Using a t-table with degrees of freedom = n1 + n2 - 2 = 68 and a significance level of 0.01, we find the critical value to be ±2.635.

Since our calculated t-value of 2.85 is greater than the critical value of 2.635, we reject the null hypothesis and 
conclude that there is sufficient evidence to suggest that the population means for the two groups are not equal.
'''
```

```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.

'''
A scatter plot is a type of chart that displays the relationship between two numerical variables. It is a graphical representation 
of a set of data points, where each point represents the values of two variables. One variable is plotted on the horizontal axis and 
the other on the vertical axis, and the location of each point on the plot corresponds to the values of the two variables for that data point.
'''

import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

'''
The subplot() function in Matplotlib is used to create multiple plots within a single figure. It takes three arguments, which specify the number of rows, 
the number of columns, and the plot number. By specifying the plot number, it is possible to create multiple plots in a grid layout.
'''

import numpy as np
import matplotlib.pyplot as plt

# Define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot each line plot in a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles and labels for each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

axs[0, 0].set_xlabel('X')
axs[0, 1].set_xlabel('X')
axs[1, 0].set_xlabel('X')
axs[1, 1].set_xlabel('X')

axs[0, 0].set_ylabel('Y')
axs[0, 1].set_ylabel('Y')
axs[1, 0].set_ylabel('Y')
axs[1, 1].set_ylabel('Y')

# Display the figure
plt.show()
```

```python
#Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

'''
A bar plot is a type of chart that represents categorical data with rectangular bars, where the length or height 
of each bar corresponds to the value of a specific category. The bars can be arranged either vertically or horizontally, 
depending on the orientation of the chart. Bar plots are also sometimes referred to as bar charts or column charts.

Bar plots are used to display and compare the values of different categories. They are particularly useful 
for visualizing the frequency, count, or percentage of categorical data, and for highlighting differences or 
similarities between categories. Bar plots can be used to display data in various fields, including statistics, business, economics, and social sciences.
'''

plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

```python
#Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
'''
A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution. It summarizes the distribution 
of a set of numerical data by dividing the data into quartiles, and displaying the median, quartiles, and outliers of the data set in a compact form.

Box plots are used to visually identify the spread, central tendency, and skewness of a dataset. They are particularly useful when comparing the distributions 
of multiple datasets, as box plots allow for easy comparisons of the median, range, and variability of each dataset. Box plots can also be used to detect 
potential outliers or unusual values in the dataset.
'''

plt.boxplot([box1, box2])
plt.show()
```

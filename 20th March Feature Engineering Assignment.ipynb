```python
#Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
'''
Nominal encoding, also known as categorical encoding, is a technique used to represent categorical variables as numerical values. 
It assigns unique integers or codes to each category, without any specific ordering or ranking implied. 
Nominal encoding is particularly useful when working with categorical variables that have no inherent order or hierarchy.

One common example of nominal encoding is one-hot encoding. 
In one-hot encoding, each category is transformed into a binary vector representation, where each bit represents 
the presence or absence of a specific category. 

Let's consider a real-world scenario to illustrate how one-hot encoding can be used:

Scenario: Customer Segmentation

Suppose you are working with a dataset containing customer information for an e-commerce company. 
One of the variables in the dataset is "Country," which represents the country of residence for each customer. 
The country variable has categorical values such as "USA," "UK," "Germany," and "France."

To utilize this categorical variable in a machine learning model for customer segmentation,
you can apply nominal encoding, specifically one-hot encoding, to convert the "Country" variable into numerical representations. 
'''

import pandas as pd
from sklearn.preprocessing  import OneHotEncoder as OHD
data = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4, 5],
    'Country': ['USA', 'UK', 'Germany', 'UK', 'France']
})
encoder = OHD()
encoded = encoder.fit_transform(data[['Country']])
encoded_data = pd.DataFrame(encoded.toarray(), columns = encoder.get_feature_names_out())
pd.concat([data, encoded_data], axis=1)
```
